- Quel est le prix moyen après réduction ?
- Quel est le pourcentage de réduction moyen ?
- Quelle est la note (rating) moyenne de tous les produits ?
- Quel est le produit le moins cher ? Le plus cher ?

```python
%run E:/DATA_ANALYST_KAGGLE/amazon-data-analyst/common/utils.ipynb
df = read_data("E:/DATA_ANALYST_KAGGLE/amazon-data-analyst/data/amazon.csv")
df = cleaning_data(df)
df = extract_main_category(df)
print(f"length of dataframe => {len(df)}")
```

```python
price_average_after_discount = df["discounted_price"].mean().round(2)
average_price_after_discount = price_average_after_discount
print(f"Prix moyenne après réduction => {average_price_after_discount}")
```

```python
discount_percentage_average = df["discount_percentage"].mean().round(2)
print(f"Pourcentage de réduction moyenne => {discount_percentage_average}%")
```

```python
rate_average_for_all_products = df["rating"].mean().round(2)
print(f"Note moyenne de tous les produits => {rate_average_for_all_products}")
```

```python
more_expensive_idx = df["discounted_price"].idxmax()
less_expensive_idx = df["discounted_price"].idxmin()

most_expensive_row = df.loc[more_expensive_idx]
least_expensive_row = df.loc[less_expensive_idx]

print(f"Le produit le plus cher => avec le prix {most_expensive_row['discounted_price']} ₹")
print(f"Le produit le moins cher => avec le prix {least_expensive_row['discounted_price']} ₹")
```
